```python
import pandas as pd
#import nlp library for positivity
import nltk
from nltk.sentiment.vader import SentimentIntensityAnalyzer
```

```python
narratives = (pd.read_csv("narratives/csv/narratives.csv")
              .drop_duplicates(subset="id")
              .reset_index(drop=True))
sia = SentimentIntensityAnalyzer()
narratives["polarity"] = [sia.polarity_scores(t)["compound"] for t in narratives["text"]]
narratives = narratives.drop(columns=["text", "polarity"])
narratives.to_csv("narratives/csv/info.csv", index=False)
```
